```python
import pandas as pd

#need glob to search the files in folder
from glob import glob
```

load all pickle files 

```python
#search in the directory to load all pkl files matches the given names and exclud * from the match
df_list= sorted(glob('../data/df_payments_*.pkl'))
#print the list of files
df_list
```

```python
#concate the list of pickle files to one dataframe
df_all_years = pd.concat((pd.read_pickle(file)
          for file in df_list), ignore_index= True)
```
